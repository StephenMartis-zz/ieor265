```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import KFold

df = pd.read_csv('winequality-red.csv', sep=';')
predictors=list(df)
predictors.remove('quality')
x = df[predictors].as_matrix()
y = df['quality'].as_matrix()

def ols(x, y):
    #Fit the model
    linreg = linear_model.LinearRegression(normalize=True)
    linreg.fit(x, y)
    y_pred = linreg.predict(x)
    
    #Return the result
    ret = [linreg.intercept_]
    ret.extend(linreg.coef_)
    return ret

def ridge(x, y, mu):
    #Fit the model
    reg = linear_model.Ridge(alpha=mu, normalize=True)
    reg.fit(x, y)
    y_pred = reg.predict(x)
    
    #Return the result
    #rss = sum((y_pred-y)**2)
    ret = [reg.intercept_]
    ret.extend(reg.coef_)
    return ret

def lasso(x, y, mu):
    #Fit the model
    reg = linear_model.Lasso(alpha=mu, normalize=True, max_iter=1e5)
    reg.fit(x, y)
    y_pred = reg.predict(x)
    
    #Return the result
    ret = [reg.intercept_]
    ret.extend(reg.coef_)
    return ret

def ede(x, y, mu, d):
    #Compute SVD and Pi
    U, s, V = np.linalg.svd(x)
    V2 = np.transpose(V)[:, d:10]
    Pi = mu*np.dot(V2, np.transpose(V2))
    
    #Append ones and zeros
    x = np.insert(x, 0, 1, axis=1)
    y = np.insert(y, len(y), np.zeros(len(Pi)))
    Pi = np.insert(Pi, 0, 0, axis=1)
    
    A = np.append(x,Pi,axis=0)
    
    #OLS fit
    reg = np.linalg.lstsq(A, y)
    
    #Return the result
    return reg[0]  

#Checked the ede limits mu=0 and d=10 yield OLS
ref = ols(x, y)
test1 = ede(x, y, 0, 6)
test2 = ede(x, y, 0.1, 10)

#5-fold cross-validation
mu = [0.01, 0.05, 0.1, 0.5, 1, 2, 3.16, 5, 10, 100]
#need different parameter range for lasso
lassomu = [0.00001, 0.00005, 0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1]
dim = range(11)
sig_lasso = []
sig_ridge = []
sig_ede = []
kf = KFold(5)

#lasso cross-validation
for m in lassomu:
    stemp = 0
    for k, (train, test) in enumerate(kf.split(x, y)):
        temp = lasso(x[train],y[train], m)
        for j in test:
            stemp = stemp+(y[j]-np.dot(x[j],temp[1:])-temp[0])**2
    #normalize by number of measurements to get average error
    sig_lasso.extend([stemp/np.size(x)])

#ridge cross-validation
for m in mu:
    stemp = 0
    for k, (train, test) in enumerate(kf.split(x, y)):
        temp = ridge(x[train],y[train], m)
        for j in test:
            stemp = stemp+(y[j]-np.dot(x[j],temp[1:])-temp[0])**2
    sig_ridge.extend([stemp/np.size(x)])

#ede cross-validation
for i in dim:
    dtemp = []
    for m in mu:
        stemp = 0
        for k, (train, test) in enumerate(kf.split(x, y)):
            temp = ede(x[train], y[train], m, i)
            for j in test:
                stemp = stemp+(y[j]-np.dot(x[j],temp[1:])-temp[0])**2
        dtemp.extend([stemp/np.size(x)])
    sig_ede.append(dtemp)

```

```python
#plot lasso cv
plt.plot(np.log10(lassomu),sig_lasso)
plt.xlabel('log($\mu$)')
plt.ylabel('CV Error')
plt.title('Lasso 5-fold CV')
plt.savefig('lasso.pdf')
plt.close()

#plot ridge cv
plt.plot(np.log10(mu),sig_ridge)
plt.xlabel('log($\mu$)')
plt.ylabel('CV Error')
plt.title('Ridge 5-fold CV')
plt.savefig('ridge.pdf')
plt.close()

#plot ede cv
#Exclude d=10 case (this is simple OLS)
for i in range(len(sig_ede)-1):
    plt.plot(np.log10(mu),sig_ede[i], label='d='+str(i))
plt.xlabel('log($\mu$)')
plt.ylabel('CV Error')
plt.title('EDE 5-fold CV')
plt.legend(loc='center left')
plt.savefig('ede.pdf')
plt.close()

#blow up d=7 and d=9 plots
plt.plot(np.log10(mu),sig_ede[7], label='d=7')
plt.plot(np.log10(mu),sig_ede[9], label='d=9')
plt.xlabel('log($\mu$)')
plt.ylabel('CV Error')
plt.title('EDE 5-fold CV (subplot)')
plt.legend(loc='upper left')
plt.savefig('ede2.pdf', bbox_inches='tight')
plt.close()
```

```python
lmu = lassomu[sig_lasso.index(min(sig_lasso))]
rmu = mu[sig_ridge.index(min(sig_ridge))]
emu = mu[sig_ede[7].index(min(sig_ede[7]))]
ed = 7
min_err = ['n/a', str(min(sig_lasso)), str(min(sig_ridge)), str(min(sig_ede[7]))]
predictors=list(df)
predictors.remove('quality')
predictors.insert(0,'intercept')

o = pd.DataFrame([ols(x, y)], columns = predictors, index=['OLS'])
l = pd.DataFrame([lasso(x, y, lmu)], columns = predictors, index=['Lasso'])
r = pd.DataFrame([ridge(x, y, rmu)], columns = predictors, index=['Ridge'])
e = pd.DataFrame([ede(x, y, emu, ed)], columns = predictors, index=['EDE'])

out = o.append([l,r,e])
out['min cv error'] = pd.Series(min_err, index=out.index)
```
